```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df1= pd.read_csv('인천광역시_하루평균근로시간_2021년.csv', encoding= 'euc-kr')
```

```python
df2= pd.read_csv('인천광역시_월 평균 가구소득_2021년.csv', encoding= 'euc-kr')
```

```python
df1
```

```python
df3= pd.merge(df1, df2)
```

```python
df3
```

```python
df3.set_index('군구', inplace= True)
```

```python
df3.describe()
```

```python
df3.to_csv('군구별_근로시간과 가구소득')
```

```python
pd.read_csv('군구별_근로시간과 가구소득', index_col=0)
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'Malgun Gothic'
df3[['1시간 이상 3시간 미만', '50만원 미만']].plot(kind= 'bar')
plt.title('군/구별 근로시간과 가구별 소득')
plt.show()
```

```python
plt.figure(figsize= (10, 8))
sb.heatmap(data= df3[['1시간 이상 3시간 미만', '3시간 이상 5시간 미만', '5시간 이상 7시간 미만', '7시간 이상 9시간 미만', '9시간 이상 12시간 미만', '12시간 이상', '50만원 미만', '50만원 이상 - 100만원 미만', '100만원 이상 - 200만원 미만', '200만원 이상 - 300만원 미만', '300만원 이상 - 400만원 미만', '400만원 이상 - 500만원 미만', '500만원 이상 - 600만원 미만', '600만원 이상 - 700만원 미만', '700만원 이상 - 800만원 미만', '800만원 이상']].corr(),
           cmap= 'RdPu', annot= True, fmt= '.1f', linewidths= 0.8)
plt.show()
```

```python
plt.figure(figsize= (10, 8))
sb.heatmap(data= df3 , cmap= 'RdPu', annot= True, fmt= '.1f', linewidths= 0.8)
plt.show()
```
